# Exploratory Data Analaysis
## Principal Component Analysis

```python
# Imports
import sys, os
sys.path.insert(0, '../src')
from skrub import TableReport
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scikit_na as na
from pathlib import Path
import preprocess
import explore
```

```python
# Load datasets
df_im, df_cl = explore.load_data()
```

```python
# Pre-Process
df_im_vis = preprocess.clean_im(df_im, verbose=False)
df_cl_vis = preprocess.clean_cl(df_cl, verbose=False)

id_cols = ['Patient', 'Timepoint', 'Date']
```

# PCA - Timepoint 1 to 5

```python
im_pca_store = explore.pca_per_timepoint(
    df_im_vis,
    timepoints=[1, 2, 3, 4, 5],
    ex_cols=id_cols,
    name='Immunological',
    ncomp=10,
)
```

## PCA - Timepoint 1 to 5 
### Colored by Clinical Variables

```python
cl_color_configs = [
    ('gender',            'categorical', 'mako'),
    ('pain_scale',        'continuous',  'mako'),
    ('diagnosis',         'categorical', 'tab20'),
]

explore.pca_colored(
    im_pca_store,
    timepoints=[1, 2, 3, 4, 5],
    color_configs=cl_color_configs,
    name='Immunological Dataset ',
    color_source_df=df_cl_vis,
)
```

# Trajectory PCA - Timepoint 1 to 3

```python
print('\nStep 2d: Trajectory PCA T1-T2, T2-T3, T1-T3 (immunological)')
_mako5 = sns.color_palette('mako', 5)
im_pairs = [
    (1, 2, _mako5[2], 'T1 -> T2'),
    (2, 3, _mako5[4], 'T2 -> T3'),
    (1, 3, _mako5[4], 'T1 -> T3'),
]

explore.trajectory_pca_im(
    df_im_vis,
    pairs=im_pairs,
    ex_cols=id_cols,
    ncomp=10,
)
```
